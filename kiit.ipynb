```python
from langchain_core.documents import Document

print(type(wiki_docs1), type(wiki_docs2))

wiki_docs1 = list(wiki_docs1)
wiki_docs2 = list(wiki_docs2)
web_docs = list(web_docs)
text_docs = list(text_docs)
csv_docs = list(csv_docs)

combined_dataset = wiki_docs1 + wiki_docs2 + web_docs + text_docs + csv_docs

print("Total combined documents:", len(combined_dataset))
for i, doc in enumerate(combined_dataset[:3]):
    print("Document:", i+1)
```

```python
from langchain_core.documents import Document

fixed_dataset = []

for doc in combined_dataset:

    # If already a Document object
    if isinstance(doc, Document):
        fixed_dataset.append(doc)

    # If tuple/list with at least 2 values
    elif isinstance(doc, (tuple, list)) and len(doc) >= 2:
        fixed_dataset.append(
            Document(
                page_content=str(doc[1]),
                metadata={"source": str(doc[0])}
            )
        )

    # If plain string
    elif isinstance(doc, str):
        fixed_dataset.append(
            Document(
                page_content=doc,
                metadata={"source": "string_data"}
            )
        )

    # Any other datatype
    else:
        fixed_dataset.append(
            Document(
                page_content=str(doc),
                metadata={"source": "unknown"}
            )
        )

combined_dataset = fixed_dataset

print("Fixed Dataset Size:", len(combined_dataset))

# Test output
for i, doc in enumerate(combined_dataset[:3]):
    print("\nDocument:", i + 1)
    print("Metadata:", doc.metadata)
    print("Content:", doc.page_content[:300])
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 500,
    chunk_overlap = 50
)

split_documents = text_splitter.split_documents(combined_dataset)

print(len(split_documents))
print(split_documents)
```

```python
import os
os.environ["TRANSFORMERS_NO_TF"] = "1"

from sentence_transformers import SentenceTransformer
import numpy as np
import faiss


embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

# Convert documents into list of strings
texts = [doc.page_content for doc in split_documents]


embeddings = embedding_model.encode(
    texts,
    batch_size=32,
    show_progress_bar=True,
    convert_to_numpy=True # FAISS accepts only numpy vector array
)

# L2 Normalization makes the magnitude of embeddings(vector) equal to 1, therefore its easy for cosine simmilarity
dimension = embeddings.shape[1]
faiss.normalize_L2(embeddings)

index = faiss.IndexFlatIP(dimension)  # Inner Product = Cosine similarity
# After L2 Normalization it stores all the vectors one by one to "index"
index.add(embeddings)

# It prints how many vectors are currently stored in the FAISS index.
print(f"FAISS index created with {index.ntotal} vectors")


def search(query, top_k=5):
    query_embedding = embedding_model.encode([query], convert_to_numpy=True)
    faiss.normalize_L2(query_embedding)
    # scores stores:- Similarity values between the query and each matched document
    # indices stores:- Indices tell us index of documents to use; those documents are combined to help generate the answer.
    scores, indices = index.search(query_embedding, top_k)

    # Those indices are used to fetch the relevant texts and store them (with scores) inside results.
    results = []
    for i, idx in enumerate(indices[0]):
        results.append({
            "score": float(scores[0][i]),
            "text": texts[idx]
        })
    return results
```

```python
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings

embedding = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

vectorstore = FAISS.from_documents(
    combined_dataset,
    embedding
)

print("FAISS Vector Store Created Successfully")
```

```python
# 1. Get the text from all documents used in the index
# Replace 'split_docs' with whatever variable holds your final chunks
full_text_data = [doc.page_content.replace("\n", " ") for doc in split_documents]

# 2. Save them to a NEW file called 'kiit_full_dataset.txt'
with open("kiit_full_dataset.txt", "w", encoding="utf-8") as f:
    for line in full_text_data:
        f.write(line + "\n")

# 3. Save your FAISS index
import faiss
faiss.write_index(index, "kiit_faiss.index")

print(f"Success! Saved {len(full_text_data)} chunks to kiit_full_dataset.txt")
```

```python
query = "Does KIIT offer law programs ?"
results = vectorstore.similarity_search(query, k=3)

for i, doc in enumerate(results):
    print(f"\nResult {i+1}:\n{doc.page_content}\n")
```
